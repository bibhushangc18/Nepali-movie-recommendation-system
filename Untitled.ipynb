```python
import pandas as pd
import json

# Load your JSON data
with open('nepali-movies.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# Convert to DataFrame
df = pd.DataFrame(data)

# Check shape and first rows
print("Total movies:", df.shape[0])
print("Columns:", df.columns.tolist())
df.head()
```

```python
import matplotlib.pyplot as plt

print("=== MISSING VALUES ===")
print(df.isnull().sum())
print()
print("=== DATA TYPES ===")
print(df.dtypes)
print()
print("=== BASIC STATS ===")
print(df.describe()
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Genre distribution
df['genre'].value_counts().head(10).plot(kind='bar', ax=axes[0], 
                                          color='steelblue', edgecolor='black')
axes[0].set_title('Top 10 Genres')
axes[0].set_xlabel('Genre')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

# Rating distribution
df['rating'].hist(bins=20, ax=axes[1], color='orange', edgecolor='black')
axes[1].set_title('Rating Distribution')
axes[1].set_xlabel('Rating')
axes[1].set_ylabel('Count')

# Movies per year — fix: convert year to numeric first
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['year'].value_counts().sort_index().plot(kind='line', ax=axes[2], color='green')
axes[2].set_title('Movies Per Year')
axes[2].set_xlabel('Year')
axes[2].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Clean runtime column
df['runtime'] = df['runtime'].str.replace(' min', '').str.strip()
df['runtime'] = pd.to_numeric(df['runtime'], errors='coerce')

# Fill missing values
df['plot']    = df['plot'].fillna('')
df['genre']   = df['genre'].fillna('Unknown')
df['rating']  = df['rating'].fillna(df['rating'].mean())
df['runtime'] = df['runtime'].fillna(df['runtime'].mean())

print("=== AFTER CLEANING - Missing Values ===")
print(df.isnull().sum())
print()
df.head()
```

```python
# Combine genre + plot into one 'tags' column
df['tags'] = df['genre'].str.lower() + ' ' + df['plot'].str.lower()

# Clean tags
df['tags'] = df['tags'].str.replace('  ', ' ').str.strip()

# Reset index
df = df.reset_index(drop=True)

print("=== TAGS SAMPLE ===")
print(df[['title', 'tags']].head(3).to_string())
```
